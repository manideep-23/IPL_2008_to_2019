# Highest score by players in ipl in single match more than 100

```python
ipl_rohit=pd.read_csv('deliveries.csv')
d = {'batsman_runs': ['sum']}

ipl_rohit2=ipl_rohit.groupby(['match_id','batsman']).agg(d)
ipl_rohit2.columns = ['_'.join(col) for col in ipl_rohit2.columns.values]

#ipl_rohit2=ipl_rohit.groupby(['over','match_id',]).agg('batsman_runs':lambda x: (x>=20).sum())
ipl_rohit_filter=ipl_rohit2[ipl_rohit2.batsman_runs_sum>=100]
ipl_rohit_filter.sort_values(by=['batsman_runs_sum'],ascending=False)
```

# who has more centuries in ipl?

# The player who hit more than 20 runs in a single over?
